```python
# importing libraries
# the utilities file has most of the coding, that's where I have done the dirty work to make this neat
from utilities import *

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
# file paths
catalogs_file_path = r"data\DMW_Clicks_Catalogs_2.xlsx"
clicks_file_path = r"data\DMW_Clicks_Catalogs_2.xlsx"
```

```python
# parameters
# you can change the frequency to D - daily, monthy for monthly output, to yearly as it pleases
freq = "M"
```

```python
# read catalogs data set
dmw_catalogs = pd.read_excel(catalogs_file_path, sheet_name="Catalogs_03_04", parse_dates=["Date"], index_col=0)
dmw_catalogs.columns = dmw_catalogs.columns.str.lower().str.replace(" ", "_")
dmw_catalogs.index.name = dmw_catalogs.index.name.lower()
dmw_catalogs_ts = dmw_catalogs.resample(freq).sum()
# dmw_catalogs_ts.head(2)
```

```python
# read clicks data
dmw_clicks = pd.read_excel(clicks_file_path, sheet_name="Clicks_03_04", parse_dates=["Date"], index_col=0)
dmw_clicks.columns = dmw_clicks.columns.str.lower().str.replace(" ", "_")
dmw_clicks.index.name = dmw_clicks.index.name.lower()
dmw_clicks_ts = dmw_clicks.resample(freq).sum()
# dmw_clicks_ts.head(2)
```

```python
# merge catalogs data and clicks data
dmw_catalogs_clicks_ts = dmw_catalogs_ts.merge(dmw_clicks_ts, left_index=True, right_index=True, how="right")
dmw_catalogs_clicks_ts
```

# In Feb, 2013,and Feb, Mar and April, 2014, there were no catalogs mailed but we had clicks 

```python
 # correlation of the catalogs mailed to the number of clicks
corr, p = stats.pearsonr(dmw_catalogs_clicks_ts['cats_mailed'], dmw_catalogs_clicks_ts['total'])
print(corr, p)
```
